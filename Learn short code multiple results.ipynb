```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Kutengeneza data kwa kutumia NumPy (Hapa namba hazitapotea)
miezi = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug"]

data_mauzo = {
    "Mwezi": miezi,
    "Shati": np.array([120, 135, 150, 145, 190, 210, 240, 260]),
    "Suruali": np.array([80, 95, 110, 105, 130, 145, 170, 190]),
    "Viatu": np.array([50, 65, 60, 75, 85, 90, 110, 120]),
    "Kofia": np.array([30, 45, 40, 60, 55, 75, 70, 90])
}

# 2. Convert data kuwa DataFrame ya Pandas
df = pd.DataFrame(data_mauzo)
bidhaa=df.columns.drop("Mwezi")
colour=["blue","yellow","green","red"]
fig,axes=plt.subplots(2,2,figsize=(10,10))
axes=axes.flatten()
for index, kitu in enumerate(bidhaa):
  sns.lineplot(data=df,x="Mwezi",y=kitu,color=colour[index],markers="x",ax=axes[index])

  axes[index].set_title(f"Mauzo: {kitu}")
  axes[index].set_xlabel("Mwezi")
  axes[index].set_ylabel(f"{kitu}")
plt.tight_layout()
plt.show()
```
